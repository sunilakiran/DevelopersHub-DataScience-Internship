```python
# Core data manipulation
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 2)
plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='whitegrid', palette='Set2')

print('All libraries imported successfully!')
```

```python
# Load Iris dataset from seaborn (downloads automatically)
df = sns.load_dataset('iris')

print('Dataset loaded successfully!')
print(f'Shape: {df.shape[0]} rows x {df.shape[1]} columns')
```

```python
palette = {'setosa': '#2ecc71', 'versicolor': '#3498db', 'virginica': '#e74c3c'}
species_list = df['species'].unique()
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
feature_labels = ['Sepal Length (cm)', 'Sepal Width (cm)', 'Petal Length (cm)', 'Petal Width (cm)']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Scatter Plots: Feature Relationships by Species', fontsize=16, fontweight='bold', y=1.02)

scatter_configs = [
    ('sepal_length', 'sepal_width',  'Sepal Length (cm)', 'Sepal Width (cm)',  'Sepal Length vs Sepal Width'),
    ('petal_length', 'petal_width',  'Petal Length (cm)', 'Petal Width (cm)',  'Petal Length vs Petal Width'),
    ('sepal_length', 'petal_length', 'Sepal Length (cm)', 'Petal Length (cm)', 'Sepal Length vs Petal Length'),
]

for ax, (x, y, xl, yl, title) in zip(axes, scatter_configs):
    for sp in species_list:
        s = df[df['species'] == sp]
        ax.scatter(s[x], s[y], label=sp.capitalize(),
                   color=palette[sp], alpha=0.8, edgecolors='white', s=70)
    ax.set_xlabel(xl, fontsize=12)
    ax.set_ylabel(yl, fontsize=12)
    ax.set_title(title, fontsize=13)
    ax.legend()

plt.tight_layout()
plt.savefig('scatter_plots.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal Length vs Petal Width shows the clearest separation between species!')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Scatter Plots: Feature Relationships by Species', fontsize=16, fontweight='bold', y=1.02)

scatter_configs = [
    ('sepal_length', 'sepal_width',  'Sepal Length (cm)', 'Sepal Width (cm)',  'Sepal Length vs Sepal Width'),
    ('petal_length', 'petal_width',  'Petal Length (cm)', 'Petal Width (cm)',  'Petal Length vs Petal Width'),
    ('sepal_length', 'petal_length', 'Sepal Length (cm)', 'Petal Length (cm)', 'Sepal Length vs Petal Length'),
]

for ax, (x, y, xl, yl, title) in zip(axes, scatter_configs):
    for sp in species_list:
        s = df[df['species'] == sp]
        ax.scatter(s[x], s[y], label=sp.capitalize(),
                   color=palette[sp], alpha=0.8, edgecolors='white', s=70)
    ax.set_xlabel(xl, fontsize=12)
    ax.set_ylabel(yl, fontsize=12)
    ax.set_title(title, fontsize=13)
    ax.legend()

plt.tight_layout()
plt.savefig('scatter_plots.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal Length vs Petal Width shows the clearest separation between species!')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Histograms: Feature Distributions by Species', fontsize=16, fontweight='bold')
axes = axes.flatten()

for i, (feature, label) in enumerate(zip(features, feature_labels)):
    for sp in species_list:
        s = df[df['species'] == sp]
        axes[i].hist(s[feature], bins=15, alpha=0.6,
                     label=sp.capitalize(), color=palette[sp], edgecolor='white')
    axes[i].set_xlabel(label, fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)
    axes[i].set_title(f'Distribution of {label}', fontsize=13)
    axes[i].legend()

plt.tight_layout()
plt.savefig('histograms.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Setosa has very distinct, compact petal distributions.')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Box Plots: Outliers and Spread by Species', fontsize=16, fontweight='bold')
axes = axes.flatten()

for i, (feature, label) in enumerate(zip(features, feature_labels)):
    sns.boxplot(
        data=df, x='species', y=feature, hue='species',
        palette=palette, ax=axes[i], legend=False,
        linewidth=1.5, flierprops=dict(marker='o', markersize=5, alpha=0.6)
    )
    axes[i].set_xlabel('Species', fontsize=12)
    axes[i].set_ylabel(label, fontsize=12)
    axes[i].set_title(f'{label} by Species', fontsize=13)

plt.tight_layout()
plt.savefig('boxplots.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Minor outliers in Setosa sepal width. Virginica has the largest measurements overall.')
```

```python
pair_plot = sns.pairplot(
    df, hue='species', palette=palette,
    diag_kind='kde', plot_kws={'alpha': 0.7, 'edgecolor': 'white'},
    height=2.5
)
pair_plot.fig.suptitle('Pair Plot: All Feature Combinations', fontsize=16, fontweight='bold', y=1.02)
plt.savefig('pairplot.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal features provide the strongest visual separation between all 3 species.')
```

```python
corr_matrix = df[features].corr()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
    center=0, vmin=-1, vmax=1, square=True,
    linewidths=0.5, cbar_kws={'shrink': 0.8}, ax=ax
)
ax.set_title('Feature Correlation Heatmap', fontsize=15, fontweight='bold', pad=15)
ax.set_xticklabels([l.replace('_', ' ').title() for l in features], rotation=30, ha='right')
ax.set_yticklabels([l.replace('_', ' ').title() for l in features], rotation=0)
plt.tight_layout()
plt.savefig('heatmap.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal Length and Petal Width are highly correlated (r = 0.96).')
```

```python
print('=' * 65)
print('         KEY STATISTICAL FINDINGS PER SPECIES')
print('=' * 65)

for sp in species_list:
    subset = df[df['species'] == sp]
    print(f'\n{sp.upper()}')
    for feat, label in zip(features, feature_labels):
        print(f'  {label:<22}: Mean={subset[feat].mean():.2f} cm,  Std={subset[feat].std():.2f}')

corr = df[features].corr()
print('\n' + '=' * 65)
print('\nCORRELATION HIGHLIGHTS:')
print(f'  Petal Length <-> Petal Width  : r = {corr.loc["petal_length","petal_width"]:.2f}  (very strong positive)')
print(f'  Sepal Length <-> Petal Length : r = {corr.loc["sepal_length","petal_length"]:.2f}  (strong positive)')
print(f'  Sepal Width  <-> Petal Length : r = {corr.loc["sepal_width","petal_length"]:.2f}  (negative)')
```
